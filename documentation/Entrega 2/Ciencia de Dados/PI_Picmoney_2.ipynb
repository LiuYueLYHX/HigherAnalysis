ETAPA 1 - Selecionar os Dados

```python
import pandas as pd

transacoes = pd.read_excel("PicMoney-Base_de_Transa__es_-_Cupons_Capturados-100000 linhas (1).xlsx")
pedestres = pd.read_excel("PicMoney-Base_Simulada_-_Pedestres_Av__Paulista-100000 linhas (1).xlsx")
lojas = pd.read_excel("PicMoney-Massa_de_Teste_com_Lojas_e_Valores-10000 linhas (1).xlsx")
players = pd.read_excel("PicMoney-Base_Cadastral_de_Players-10_000 linhas (1).xlsx")
```

```python
for nome, base in zip(['Transacoes', 'Pedestres', 'Lojas', 'Players'], [transacoes, pedestres, lojas, players]):
    print(f"\n{nome}: {base.shape[0]} linhas e {base.shape[1]} colunas")
    print(base.columns)
```

ETAPA 2 — Limpar e Uniformizar os Dados

```python
def limpar_base(df):
    df = df.copy()
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    df = df.drop_duplicates()

    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].astype(str).str.strip()

    for col in df.columns:
        if "data" in col:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    return df
```

```python
transacoes.head()
#teste de limpeza
```

```python
transacoes = limpar_base(transacoes)
pedestres = limpar_base(pedestres)
lojas = limpar_base(lojas)
players = limpar_base(players)
```

ETAPA 3 — Derivar Dados

```python
if "quantidade" in transacoes.columns and "valor_unitario" in transacoes.columns:
    transacoes["valor_total"] = transacoes["quantidade"] * transacoes["valor_unitario"]
```

```python
if "idade" in players.columns:
    players["faixa_etaria"] = pd.cut(
        players["idade"],
        bins=[0, 18, 30, 45, 60, 120],
        labels=["0-18", "19-30", "31-45", "46-60", "60+"]
    )
```

ETAPA 4 — Integrar os Dados

```python
print("Colunas Transações:\n", transacoes.columns.tolist(), "\n")
print("Colunas Pedestres:\n", pedestres.columns.tolist(), "\n")
print("Colunas Lojas:\n", lojas.columns.tolist(), "\n")
print("Colunas Players:\n", players.columns.tolist(), "\n")
```

```python
import pandas as pd

# Cópias de trabalho
t = transacoes.copy()
p = players.copy()
pdst = pedestres.copy()
lj = lojas.copy()

# Normalização de celular (mantém últimos 11 dígitos do Brasil)
def normalizar_cel(s):
    return (s.astype(str)
              .str.replace(r"\D", "", regex=True)
              .str[-11:])
```

```python
def normalizar_texto(s):
    s = s.astype(str).str.strip().str.lower()
    s = s.str.normalize("NFKD").str.encode("ascii", "ignore").str.decode("utf-8")
    s = s.str.replace(r"\s+", " ", regex=True)
    return s
```

```python
if "celular" in t.columns:   t["celular"] = normalizar_cel(t["celular"])
if "celular" in p.columns:   p["celular"] = normalizar_cel(p["celular"])
if "celular" in pdst.columns: pdst["celular"] = normalizar_cel(pdst["celular"])
if "numero_celular" in lj.columns:
    lj["celular"] = normalizar_cel(lj["numero_celular"])
```

```python
if "nome_estabelecimento" in t.columns:
    t["nome_estab_norm"] = normalizar_texto(t["nome_estabelecimento"])
if "nome_loja" in lj.columns:
    lj["nome_loja_norm"] = normalizar_texto(lj["nome_loja"])
```

```python
print("Exemplo de celulares e nomes normalizados:")
print(t[["celular","nome_estabelecimento","nome_estab_norm"]].head(3))
print(lj[["nome_loja","nome_loja_norm"]].head(3))
```

Etapa 4.1 integração Transações/Players

```python
cols_p = [c for c in p.columns if c != "celular"]
base_tp = t.merge(p[["celular"] + cols_p], on="celular", how="left", suffixes=("", "_player"))

print("\n[TP] Transações + Players")
print("Linhas antes (t):", len(t), " | depois (base_tp):", len(base_tp))
print("Cobertura de players (ex.: idade):",
      base_tp["idade"].notna().mean() if "idade" in base_tp.columns else "coluna 'idade' não existe")

print("\nAmostra:")
print(base_tp.head(5))
```

ETAPA 4.2 — Integrar Transações/Lojas

```python
import pandas as pd

def normalizar_texto(s):
    s = s.astype(str).str.strip().str.lower()
    s = s.str.normalize("NFKD").str.encode("ascii", "ignore").str.decode("utf-8")
    s = s.str.replace(r"\s+", " ", regex=True)
    return s

if "nome_estabelecimento" in t.columns and "nome_estab_norm" not in t.columns:
    t["nome_estab_norm"] = normalizar_texto(t["nome_estabelecimento"])
if "nome_estab_norm" not in base_tp.columns:
    if "nome_estabelecimento" in base_tp.columns:
        base_tp["nome_estab_norm"] = normalizar_texto(base_tp["nome_estabelecimento"])
    elif "nome_estab_norm" in t.columns and len(base_tp) == len(t):
        base_tp["nome_estab_norm"] = t["nome_estab_norm"].values
    else:
        raise ValueError("Não foi possível criar 'nome_estab_norm' em base_tp. Verifique se 'nome_estabelecimento' existe.")
if "nome_loja" in lj.columns and "nome_loja_norm" not in lj.columns:
    lj["nome_loja_norm"] = normalizar_texto(lj["nome_loja"])

if "data_captura" in lj.columns:
    lj["data_captura"] = pd.to_datetime(lj["data_captura"], errors="coerce")

print("base_tp colunas checadas:")
print([c for c in base_tp.columns if "nome" in c])
print("\nExemplo base_tp:")
print(base_tp[["nome_estabelecimento","nome_estab_norm"]].head(3))

print("\nExemplo lj:")
cols_show = [c for c in ["nome_loja","nome_loja_norm","data_captura"] if c in lj.columns]
print(lj[cols_show].head(3))
```

```python
assert "nome_estab_norm" in base_tp.columns
assert "nome_loja_norm" in lj.columns

if "data_captura" in lj.columns:
    lj_ult = (
        lj.sort_values(["nome_loja_norm","data_captura"])
          .groupby("nome_loja_norm", as_index=False)
          .tail(1)
    )
else:
    lj_ult = lj.drop_duplicates(subset=["nome_loja_norm"]).copy()

cols_lj_use = [c for c in ["nome_loja_norm","tipo_cupom","tipo_loja","local_captura",
                           "latitude","longitude","endereco_loja","valor_compra","valor_cupom","data_captura"]
               if c in lj_ult.columns]

base_tpl = base_tp.merge(
    lj_ult[["nome_loja_norm"] + [c for c in cols_lj_use if c != "nome_loja_norm"]],
    left_on="nome_estab_norm",
    right_on="nome_loja_norm",
    how="left"
)

print("\n[TPL] + Lojas")
print("Linhas antes (base_tp):", len(base_tp), " | depois (base_tpl):", len(base_tpl))
match_rate = base_tpl["nome_loja_norm"].notna().mean() if "nome_loja_norm" in base_tpl.columns else 0
print("Casamentos por nome de loja:", f"{match_rate:.1%}")

if "nome_loja_norm" in base_tpl.columns:
    na_mask = base_tpl["nome_loja_norm"].isna()
    if na_mask.any():
        print("\nTop 15 estabelecimentos sem correspondência em Lojas:")
        print(base_tpl.loc[na_mask, "nome_estab_norm"].value_counts().head(15))
```

Etapa 4.3 Integrar com Pedestres

```python

pdst_work = pdst.copy()

# Montar timestamp do pedestre
pdst_work["data"] = pd.to_datetime(pdst_work["data"], errors="coerce")
if "horario" in pdst_work.columns:
    h = pd.to_timedelta(pdst_work["horario"], errors="coerce").fillna(pd.Timedelta(0))
    pdst_work["dt_pedestre"] = pdst_work["data"] + h
else:
    pdst_work["dt_pedestre"] = pdst_work["data"]

# Selecionar o registro mais recente por celular
pdst_ult = (pdst_work
            .sort_values(["celular","dt_pedestre"])
            .groupby("celular", as_index=False)
            .tail(1))

cols_pd_use = [c for c in pdst_ult.columns if c not in ["data","horario"]]

base_final = base_tpl.merge(
    pdst_ult[cols_pd_use],
    on="celular",
    how="left",
    suffixes=("", "_pedestre")
)

print("[FINAL] + Pedestres (registro mais recente por celular)")
print("Linhas antes (base_tpl):", len(base_tpl), " | depois (base_final):", len(base_final))
if "possui_app_picmoney" in base_final.columns:
    print("Cobertura pedestres (possui_app_picmoney):",
          f"{base_final['possui_app_picmoney'].notna().mean():.1%}")
print("\nAmostra:")
print(base_final.head(5))
```

ETAPA 4.4

```python
print("Checagem de nulos em campos-chave:")
for campo in ["celular", "nome_estab_norm", "nome_loja_norm"]:
    if campo in base_final.columns:
        print(f"{campo}: {base_final[campo].isna().sum()} nulos")

print("\nCobertura por tabela (amostras):")
if "idade" in base_final.columns:
    print("Players (idade):", f"{base_final['idade'].notna().mean():.1%}")
if "tipo_loja" in base_final.columns:
    print("Lojas (tipo_loja):", f"{base_final['tipo_loja'].notna().mean():.1%}")
if "dt_pedestre" in base_final.columns:
    print("Pedestres (dt_pedestre):", f"{base_final['dt_pedestre'].notna().mean():.1%}")

# Top estabelecimentos sem correspondência em Lojas (se aplicável)
if "nome_loja_norm" in base_final.columns and "nome_estab_norm" in base_final.columns:
    mask_na_loja = base_final["nome_loja_norm"].isna()
    if mask_na_loja.any():
        print("\nTop 20 estabelecimentos sem correspondência em Lojas:")
        print(base_final.loc[mask_na_loja, "nome_estab_norm"].value_counts().head(20))
```

ETAPA 5 Formatar os Dados

```python
# Seleção de colunas finais
colunas_principais = [
    # Transações
    "celular", "data", "hora", "dt_transacao", "produto",
    "nome_estabelecimento", "nome_estab_norm", "categoria_estabelecimento",
    "id_campanha", "id_cupom", "tipo_cupom", "valor_cupom", "repasse_picmoney",
    # Players
    "idade", "sexo", "cidade_residencial", "bairro_residencial",
    "cidade_trabalho", "bairro_trabalho", "cidade_escola", "bairro_escola",
    "categoria_frequentada", "faixa_etaria",
    # Lojas
    "nome_loja_norm", "tipo_loja", "local_captura", "endereco_loja",
    "latitude", "longitude", "valor_compra", "data_captura",
    # Pedestres (último evento)
    "dt_pedestre", "local", "tipo_celular", "modelo_celular",
    "possui_app_picmoney", "data_ultima_compra", "ultimo_tipo_cupom",
    "ultimo_valor_capturado", "ultimo_tipo_loja"
]

colunas_existentes = [c for c in colunas_principais if c in base_final.columns]
base_pronta = base_final[colunas_existentes].copy()

# Ajustes de tipo básicos
for c in ["valor_cupom", "valor_compra", "repasse_picmoney", "ultimo_valor_capturado"]:
    if c in base_pronta.columns:
        base_pronta[c] = pd.to_numeric(base_pronta[c], errors="coerce")

for c in ["data", "dt_transacao", "data_captura", "dt_pedestre", "data_ultima_compra"]:
    if c in base_pronta.columns:
        base_pronta[c] = pd.to_datetime(base_pronta[c], errors="coerce")

# Ordenar por tempo da transação, se existir
if "dt_transacao" in base_pronta.columns:
    base_pronta = base_pronta.sort_values("dt_transacao")

print("Base formatada: linhas =", len(base_pronta), " | colunas =", len(base_pronta.columns))
print("Colunas finais:")
print(base_pronta.columns.tolist() )

# Exportações
base_pronta.to_excel("Base_PicMoney_Preparada.xlsx", index=False)
base_pronta.to_csv("Base_PicMoney_Preparada.csv", index=False, sep=";")
print("Arquevios salvos:Base_PicMoney_Preparada.xlsx e Base_PicMoney_Preparada.csv")
```
